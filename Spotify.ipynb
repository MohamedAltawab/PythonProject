```python
# IMPORTING LIBRARIES NEEDED

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#READING CSV FILES

df_tracks=pd.read_csv('F:/DATA ANALYSIS/Projects/Spotify/tracks.CSV')
df_tracks.head()
```

```python
#CLEANING PROCESS
# CHECKING NULL VALUES

pd.isnull(df_tracks).sum()
```

```python
#SHOW DUPLICATES

df_tracks[df_tracks.duplicated(['id'], keep=False)]
```

```python
#SHOW SOME INFORMARION ABOUT TRACKS TABLE

df_tracks.info()
```

```python
#THE LOWEST TRACKS BASED ON POPULARITY

sorted_df=df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
#APPLY DESCRIPTIVE STATSISTICS

df_tracks.describe().transpose() #transpose : inverse axis
```

```python
# SELECT TOP 10 TRACKS BASED ON POPULARITY

most_popular = df_tracks.query('popularity > 90', inplace= False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
# REMOVE SPACES IN COLUMN NAMES

df_tracks.columns = df_tracks.columns.str.replace(' ', '')
```

```python
#FORMATTING "release_date" COLUMN TO DATE

df_tracks.set_index("release_date", inplace=True)
df_tracks.index= pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# ADD NEW COLUMN ( duration) = change duration_ms from ms to seconds  

df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
```

```python
#DROP THE OLD duration_ms COLUMN

df_tracks.drop("duration_ms", inplace = True, axis =1)
```

```python
#SHOW TOP ROWS FROM TRACKS TABLE

df_tracks.duration.head()
```

```python
#ANALYSIS AND VISUALIZATION  
#CREATING A HEATMAP CHART SHOWING THE CORRELATION BETWEEN ATTRIBUTES

corr_df=df_tracks.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))    #annot shows numeric values on heatmap, while fmt shows strings
heatmap=sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, linewidths=1, linecolor='black')
heatmap.set_title("Correlation HeatMap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation =90)
```

```python
# CREATING REGRESSION PLOT "loudness vs energy correlation"

sample_df = df_tracks.sample(int(0.004*len(df_tracks)))  
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness" , x ="energy", color = "c").set(title= "loudness vs energy correlation")
```

```python
# CREATING REGRESSION PLOT "popularity vs accousticness correlation"

plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y="acousticness", x = "popularity", color = "c").set(title="popularity vs accousticness correlation")
```

```python
#CREATING HISTOGRAM CHART "Number of songs per year"
# TO SHOW THE NUMBER OF SONGS CREATING EVERY YEAR

df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year    #CONVERTING RELEASE DATE TO JUST YEAR
```

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind='hist').set(title="Number of songs per year")
```

```python
#CREATING A BAR CHART "Top 5 songs by popularity" 
#SHOWING THE TOP 5 SONGS BASED ON POPULARITY

sns.set_style(style="darkgrid")
popular= df_tracks.sort_values("popularity", ascending = False).head(10)
sns.barplot(y='name', x='popularity', data= popular). set(title="Top 5 songs by popularity")
```

```python
#IMPORTING "SpotifyFeatures" CSV FILE 
df_genre= pd.read_csv(('F:/DATA ANALYSIS/Projects/Spotify/SpotifyFeatures.CSV'))
```

```python
# SHOWING TOP 5 ROWS FROM 'GENRE'
df_genre.head()
```

```python
#CREATING A BAR CHART SHOWING DURATION OF SONGS ACCORDING TO GENRES

plt.title("Duration of the songs in different genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y='genre', x='duration_ms', data= df_genre)
plt.xlabel=("Duration in milli seconds")
plt.ylabel=("Genres")
```

```python
# SHOWING TOP 5 GENRES BASED ON POPULARITY

sns.set_style(style="darkgrid")
famous= df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(y='genre', x='popularity', data= famous). set(title="Top 5 Genre by popularity")
```

```python
# CREATING A PIE CHART SHOWING TOP GENRES BASED ON POPULARITY

sns.set_style(style="darkgrid")
piee= df_genre.sort_values("popularity", ascending = False).head(10)
piee.groupby(['genre']).sum().plot(kind='pie', y='popularity')
```
